```python
# import libraries
from pathlib import Path
import csv

# open csv file
menu_filepath = Path('/Users/kwamevanleeuwen/Bootcamp/cu-nyc-fin-pt-12-2020-u-c/02-Homework/02-Python/Instructions/PyRamen/Resources/menu_data.csv')
sales_filepath = Path('/Users/kwamevanleeuwen/Bootcamp/cu-nyc-fin-pt-12-2020-u-c/02-Homework/02-Python/Instructions/PyRamen/Resources/sales_data.csv')

# initialize lists
menu = []
sales = []

with open (menu_filepath, 'r') as csvfile:

    # create iterable reader object
    csvreader = csv.reader(csvfile, delimiter = ',')
    
    # read header and append to menu
    header = next(csvreader)
    
    # append remaining csvreader rows to menu
    for row in csvreader:
        menu.append(row)

with open (sales_filepath, 'r') as csvfile:

    # create iterable reader object
    csvreader = csv.reader(csvfile, delimiter = ',')
    
    # read header and append to menu
    header = next(csvreader)
    
    # append remaining csvreader rows to menu
    for row in csvreader:
        sales.append(row)

# initialize report dictionary
report = {}

# Initialize loop variables
row_count = 0
menu_item_toggle = 0

# loop through sales list object

for row in sales:
    quantity = int(row[3])
    sales_item = row[4]
    
    # check if sales_item already included in report
    if sales_item not in report:
        report[sales_item] = {
                             "01-count": 0,
                             "02-revenue": 0,
                             "03-cogs": 0,
                             "04-profit": 0,  
                             }
    for line in menu:
        item = line[0]
        price = float(line[3])
        cost = float(line[4])
        profit = price - cost

        if item == sales_item:
            # print matching menu data
            # print(item, price, cost, profit)
            
            #toggle menu item identified
            menu_item_toggle = 1
    
            # update cumulative menu item tracker
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += quantity * price
            report[sales_item]["03-cogs"] += quantity * cost
            report[sales_item]["04-profit"] += quantity * profit
            
    # report if sales item was not found in menu list object
    if menu_item_toggle == 0:
        print(f"{sales_item} did not find a match!")
        menu_item_toggle = 0

    # increase record count
    row_count +=1

# calculate total metrics
# initialize loop variables
tot_bowls = 0
tot_revenue = 0
tot_cogs = 0
tot_profit = 0
tot_ramentype_sold = 0
max_count = 0
max_count_ramen = 0
min_count = 0
min_count_ramen = 0
max_rev = 0
max_rev_ramen = 0
min_rev = 0
min_rev_ramen = 0
max_profit = 0
max_profit_ramen = 0
min_profit = 0
min_profit_ramen = 0

for item, value in report.items():
    tot_bowls += value["01-count"]
    tot_revenue += value["02-revenue"]
    tot_cogs += value["03-cogs"]
    tot_profit += value["04-profit"]
    tot_ramentype_sold += 1
    
    if min_count == 0:
        min_count = value["01-count"]
    elif value["01-count"] < min_count:
        min_count = value["01-count"]
        min_count_ramen = item
    if value["01-count"] > max_count:
        max_count = value["01-count"]
        max_count_ramen = item
    
    if min_rev == 0:
        min_rev = value["02-revenue"]
    elif value["02-revenue"] < min_rev:
        min_rev = value["02-revenue"]
        min_rev_ramen = item
    if value["02-revenue"] > max_rev:
        max_rev = value["02-revenue"]
        max_rev_ramen = item
    
    if min_profit == 0:
        min_profit = value["04-profit"]
    elif value["04-profit"] < min_profit:
        min_profit = value["04-profit"]
        min_profit_ramen = item
    if value["04-profit"] > max_profit:
        max_profit = value["04-profit"]
        max_profit_ramen = item
    
# calculate average metrics
bowls_per_ramentype = round(tot_bowls / tot_ramentype_sold, 2)
rev_per_ramentype = round(tot_revenue / tot_ramentype_sold, 2)
cogs_per_ramentype = round(tot_cogs / tot_ramentype_sold, 2)
profit_per_ramentype = round(tot_profit / tot_ramentype_sold, 2)

# calculate filter lists
outperforming = []
underperforming = []

for item, value in report.items():
    if value["04-profit"] > profit_per_ramentype:
        outperforming.append(item)
    elif value["04-profit"] < profit_per_ramentype:
        underperforming.append(item)

# print total records in sales list objects
print(f"Total records: {row_count}\n")

# print total metrics
print("")
print(f"Total bowls: {total_bowls}")
print(f"Total revenue: ${total_revenue}")
print(f"Total cogs: ${total_cogs}")
print(f"Total profit: ${total_profit}")

# print average metrics
print("")
print(f"Total ramen types: {tot_ramentype_sold}")
print(f"Bowls sold per ramen type: {bowls_per_ramentype}")
print(f"Revenue per ramen type: ${rev_per_ramentype}")
print(f"Cogs per ramen type: ${cogs_per_ramentype}")
print(f"Profit per ramen type: ${profit_per_ramentype}")

# Min/Max stats
print("")
print(f"Most popular ramen is: {max_count_ramen} with {max_count} bowls sold")
print(f"Least popular ramen is: {min_count_ramen} with {min_count} bowls sold")
print(f"Highest revenue ramen is: {max_rev_ramen} with ${max_rev} sold")
print(f"Lowest revenue ramen is: {min_rev_ramen} with ${min_rev} sold")
print(f"Most profitable ramen is: {max_profit_ramen} with ${max_profit} in profit")
print(f"Least profitable ramen is: {min_profit_ramen} with ${min_profit} in profit")

# filter lists
print("")
print(f"Outperforming items: {outperforming}")
print(f"Underperforming items: {underperforming}")

# write out report to txt file
# set the output file path
output_path = Path('/Users/kwamevanleeuwen/Bootcamp/Repositories/python-homework/PyRamen/output.txt')
    
# open the output path as an object
with open(output_path, 'w') as file:

    file.write(f"PyRamen report\n")
    file.write("\n")

    for key, value in report.items():
        count = value['01-count']
        revenue = value['02-revenue']
        cogs = value['03-cogs']
        profit = value['04-profit']
        file.write(f"The stats for {key} are: \n")
        file.write(f"Total count: {count}\n")
        file.write(f"Total revenue: {revenue}\n")
        file.write(f"Total profit: {profit}\n")
        file.write("\n")

    file.write(f"Summary analysis report\n")
    file.write("\n")
    

    file.write(f"Total bowls: {total_bowls}\n")
    file.write(f"Total revenue: ${total_revenue}\n")
    file.write(f"Total cogs: ${total_cogs}\n")
    file.write(f"Total profit: ${total_profit}\n")
    file.write("\n")
    
    # print average metrics
    file.write(f"Total ramen types: {tot_ramentype_sold}\n")
    file.write(f"Bowls sold per ramen type: {bowls_per_ramentype}\n")
    file.write(f"Revenue per ramen type: ${rev_per_ramentype}\n")
    file.write(f"Cogs per ramen type: ${cogs_per_ramentype}\n")
    file.write(f"Profit per ramen type: ${profit_per_ramentype}\n")
    file.write("\n")
    
    # Min/Max stats
    file.write("")
    file.write(f"Most popular ramen is: {max_count_ramen} with {max_count} bowls sold\n")
    file.write(f"Least popular ramen is: {min_count_ramen} with {min_count} bowls sold\n")
    file.write(f"Highest revenue ramen is: {max_rev_ramen} with ${max_rev} sold\n")
    file.write(f"Lowest revenue ramen is: {min_rev_ramen} with ${min_rev} sold\n")
    file.write(f"Most profitable ramen is: {max_profit_ramen} with ${max_profit} in profit\n")
    file.write(f"Least profitable ramen is: {min_profit_ramen} with ${min_profit} in profit\n")
    file.write("\n")

    # filter lists
    file.write(f"Outperforming items: {outperforming}\n")
    file.write(f"Underperforming items: {underperforming}\n")
    file.write("\n")
```
